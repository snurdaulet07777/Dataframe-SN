```python
import pandas
```

Imports pandas library. Open source data analysis and manipulation tool

```python
table = pandas.read_csv('bank.csv', sep=';')
```

Reads bank.csv spreadsheet as two-dimensional data structure with labeled axes

```python
table
```

This line just prints the dataframe, which is table 2 dimensional table. Same with the csv file.

```python
table.head(10)
```

The head(10) function returns the first 10 rows. After that the returned value is printed. Usually, this function is useful for quickly testing if your object has the right type of data in it.

```python
table.tail(10)
```

The tail(10) function gets the last 10 rows. After that returned value is printed as 2 dimensional table. This functions is useful for quickly verifying data, for example, after sorting or appending rows.

```python
table.info()
```

Info method displays information such as the number of rows and columns, the total memory usage, the data type of each column, and the number of non-NaN elements.

```python
table.corr(method= 'pearson')
```

Method corr() is used to find the pairwise correlation of all columns in the dataframe. For any non-numeric data type columns in the dataframe it is ignored. Here the method of correlation is "pearson". Which is standard correlation coefficient.

```python
table.corr(method= 'kendall')
```

Same method here with different method of correlation. "kendall" is Kendall Tau correlation coefficient.

```python
table.describe()
```

Method describe() is used to view some basic statistical details like percentile, mean, std etc. of a data frame.

Description of the each row of returned table.

count: Count number of non-NA/null observations.

max: Maximum of the values in the object.

min: Minimum of the values in the object.

mean: Mean of the values.

std: Standard deviation of the observations.
